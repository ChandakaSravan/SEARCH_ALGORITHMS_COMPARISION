```python
from time import perf_counter, perf_counter_ns
import random


def linear_search(arr, ele):
    """Linear Search Implementation

    Args:
        arr (list[int]): An array to be searched for
        ele (int): Element to be searched

    Returns:
        int: If the element is found returns the index of element in arr else returns -1
    """
    idx = 0
    for i in arr:
        idx += 1
        if i == ele:
            return idx
    return -1


def binary_search(arr, low, high, k):
    """Binary Search implementation

    Args:
        arr (list[int]): Sorted array
        low (int): Lowest index in partition of Binary Search Array
        high (int): Highest index in partition of Binary Search Array
        k (int): Value to be found in array arr

    Returns:
        int: index of the searched value if found else returns -1
    """
    if high >= low:
        mid = (low + high) // 2
        mid = low + (high - low) // 2
        if arr[mid] == k:
            return mid
        elif arr[mid] > k:
            return binary_search(arr, low, mid-1, k)
        else:
            return binary_search(arr, mid+1, high, k)
    else:
        return -1

```

```python
linear = []
# sizes 100, 1000, 10000, 100000, 500000

if __name__ == "__main__":
    size = [100, 500, 1000, 5000, 10000, 50000, 100000, 150000, 200000]
    for s in size:
        arr = []
        l = len(arr)
        print("The size is : {}".format(s))
        while l < s:
            j = random.randint(0, 1000000)
            if j not in arr:
                arr.append(j)
            l = len(arr)
            lisl = []
        lt = 20
        print(len(arr))
        for i in range(lt):
            ele = random.randint(0, s-1)
            ele = arr[ele]
            start = perf_counter_ns()
            linear_search(arr, ele)
            end = perf_counter_ns()
            lisl.append(end-start)
            print("The time taken to search using linear search is : {}".format(end-start))
        suml = 0
        for idx in range(len(lisl)):
            suml += lisl[idx]
        print("Average time for linear search size {} is : {}".format(s, suml/lt))
        print("----------------------------------------------------------------------------------------------")
        linear.append(suml/lt)
```

```python
# BST
import random
from time import perf_counter, perf_counter_ns


class Node:
    """Basic Structure of a Binary Search Tree Node
    """

    def __init__(self, data) -> None:
        self.left = None
        self.right = None
        self.value = data
        self.parent = None


def insertIntoBST(root, data):
    """This is an insert function for Binary search tree it inserts the values less than or equal to root in the left of root and values greater than root into right of root.

    Args:
        root (Node): Root Node of Binary Search Tree
        data (int): Value to be inserted in binary search tree
    """
    x = root
    y = None
    z = Node(data)

    while(x != None):
        y = x
        if z.value <= x.value:
            x = x.left
        else:
            x = x.right

    if z.value <= y.value:
        y.left = z
    else:
        y.right = z
    z.parent = y


def searchTree(root, data):
    """Searches the BST for given value data and returns the index if found else returns -1

    Args:
        root (Node): Root Node of BST
        data (int): Data Value to be searched

    Returns:
        int: Index if value is found else returns -1
    """
    if root is None:
        return False
    if root.value == data:
        return True
    if root.value <= data:
        return searchTree(root.right, data)
    else:
        return searchTree(root.left, data)


def searchTreeNode(root, data):
    if root is None:
        return False
    if root.value == data:
        return root
    if root.value <= data:
        return searchTreeNode(root.right, data)
    else:
        return searchTreeNode(root.left, data)


def inOrderTraversal(root):
    if root:
        inOrderTraversal(root.left)
        print(root.value, end=" ")
        inOrderTraversal(root.right)


def preOrderTraversal(root):
    if root:
        print(root.value, end=" ")
        preOrderTraversal(root.left)
        preOrderTraversal(root.right)


def postOrderTraversal(root):
    if root:
        postOrderTraversal(root.left)
        postOrderTraversal(root.right)
        print(root.value, end=" ")


def treeMinimum(root):
    x = root
    while (x.left != None):
        x = x.left
    return x.value


def treeMaximum(root):
    x = root
    while(x.right != None):
        x = x.right
    return x.value


def BSTSort(root, arr):
    for i in range(len(arr)):
        insertIntoBST(root, arr[i])
    inOrderTraversal(root)


def treeSuccessor(root, x):
    if x.right != None:
        return treeSuccessor(x.right)
    y = x.parent
    while y != None and x == y.right:
        x = y
        y = y.parent
    return y


def DeleteFromBST(root, value):
    val = searchTreeNode(root, val)
    if val == False:
        return "Value not found in BST"
    # case 0 when node has no left and right subtrees
    if val.left == None and val.right == None:
        p = val.parent
        if p.left == val:
            p.left = None
        elif p.right == val:
            p.right = None
    # case 1 when node has a left or right subtree
    if val.left == None and val.right != None:
        p = val.parent
        if p.left == val:
            p.left = val.right
        elif p.right == val:
            p.right = val.right
    if val.left != None and val.right == None:
        p = val.parent
        if p.left == val:
            p.left = val.left
        elif p.right == val:
            p.right = val.left
    # case 2 when node has both left and right subtrees and right child has no left subtree
    if val.left != None and val.right != None:
        p = val.parent
        if val.right.left == None:
            right = val.right
            right.left = val.left
            if p.left == val:
                p.left = right
            elif p.right == val:
                p.right = right


if __name__ == "__main__":
    arr = []
    bst = []
    l = len(arr)
    size = [100, 500, 1000, 5000, 10000, 50000, 100000, 150000, 200000]
    for s in size:
        while l < s:
            j = random.randint(0, 1000000)
            if j not in arr:
                arr.append(j)
            l = len(arr)
        root = Node(arr[0])
        for i in range(1, len(arr)):
            insertIntoBST(root, arr[i])
#         if s == 1000:
#             inOrderTraversal(root)
        lisb = []
        lt = 20
        for i in range(lt):
            ele = random.randint(0, s-1)
            ele = arr[ele]
            start = perf_counter_ns()
            val = searchTree(root, ele)
            end = perf_counter_ns()
            lisb.append(end-start)
            if val == False:
                val = -1
            # print("The value searched is in index: "+val)
            print(val)
            print("The time taken to search using bst search is : {}".format(end-start))
        print("Average run time for size {} is : {}".format(s, sum(lisb)/lt))
        print("-------------------------------------------------------------------------------------")
        bst.append(sum(lisb)/lt)
```

```python
import sys


class Node():
    def __init__(self, data) -> None:
        self.value = data
        self.parent = None
        self.left = None
        self.right = None
        self.color = 1


class RBTree():
    def __init__(self) -> None:
        self.temp = Node(0)
        self.root = self.temp

    def left_rotate(self, x):
        y = x.right
        x.right = y.left
        if y.left != self.temp:
            y.left.parent = x

        y.parent = x.parent
        if x.parent == None:
            self.root = y
        elif x == x.parent.left:
            x.parent.left = y
        else:
            x.parent.right = y
        y.left = x
        x.parent = y

    def right_rotate(self, x):
        y = x.left
        x.left = y.right
        if y.right != self.temp:
            y.right.parent = x

        y.parent = x.parent
        if x.parent == None:
            self.root = y
        elif x == x.parent.right:
            x.parent.right = y
        else:
            x.parent.left = y
        y.right = x
        x.parent = y

    def insert(self, key):
        """Insertion Algorithm for a Red Black Tree followed by a insertion algorithm fixup.

        Args:
            key (int): value of new node to be inserted
        """
        node = Node(key)
        node.parent = None
        node.item = key
        node.left = self.temp
        node.right = self.temp
        node.color = 1

        y = None
        x = self.root

        while x != self.temp:
            y = x
            if node.item < x.item:
                x = x.left
            else:
                x = x.right

        node.parent = y
        if y == None:
            self.root = node
        elif node.item < y.item:
            y.left = node
        else:
            y.right = node

        if node.parent == None:
            node.color = 0
            return

        # if node.parent.parent == None:
        #     return

        self.fix_insert(node)

    # Balance the tree after insertion
    def fix_insert(self, k):
        """Tree Balancing Algorithm

        Args:
            k (Node): Node inserted during insertion Algorithm
        """
        while k.parent.color == 1:
            if k.parent == k.parent.parent.right:
                u = k.parent.parent.left
                # case 1 in rb fixup
                if u.color == 1:
                    u.color = 0
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    k = k.parent.parent
                else:
                    # case 2 in rb fixup
                    if k == k.parent.left:
                        k = k.parent
                        self.right_rotate(k)
                    # case 3 in rb fixup
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    self.left_rotate(k.parent.parent)
            else:
                u = k.parent.parent.right
                # case 1 in rb fixup
                if u.color == 1:
                    u.color = 0
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    k = k.parent.parent
                else:
                    # case 2 in rb fixup
                    if k == k.parent.right:
                        k = k.parent
                        self.left_rotate(k)
                    # case 3 in rb fixup
                    k.parent.color = 0
                    k.parent.parent.color = 1
                    self.right_rotate(k.parent.parent)
            if k == self.root:
                break
        self.root.color = 0

    def inOrderTraversal(self, root):
        if root:
            self.inOrderTraversal(root.left)
            print(root.value, end=" ")
            self.inOrderTraversal(root.right)

    def preOrderTraversal(self):
        root = self.root
        if root:
            print(root.value, end=" ")
            self.preOrderTraversal(root.left)
            self.preOrderTraversal(root.right)

    def postOrderTraversal(self):
        root = self.root
        if root:
            self.postOrderTraversal(root.left)
            self.postOrderTraversal(root.right)
            print(root.value, end=" ")

    def treeMinimum(self):
        x = self.root
        while (x.left != None):
            x = x.left
        return x.value

    def treeMaximum(self):
        x = self.root
        while(x.right != None):
            x = x.right
        return x.value

    def searchTree(self, root, data):
        """Searches the RB Tree for given value data and returns the index if found else returns -1

        Args:
            root (Node): Root Node of RB Tree
            data (int): Data Value to be searched

        Returns:
            int: Index if value is found else returns -1
        """
        if root is None:
            return False
        if root.value == data:
            return True
        if root.value <= data:
            return self.searchTree(root.right, data)
        else:
            return self.searchTree(root.left, data)


if __name__ == "__main__":
    tree = RBTree()
    arr = [55, 40, 65, 60, 75, 57, 443, 342, 787]
    for i in arr:
        tree.insert(i)
    print(tree.searchTree(tree.root, 787))
    print(tree.searchTree(tree.root, 4))
    
    arr = []
    rbt = []
    l = len(arr)
    size = [100, 500, 1000, 5000, 10000, 50000, 100000, 150000, 200000]
    for s in size:
        while l < s:
            j = random.randint(0, 1000000)
            if j not in arr:
                arr.append(j)
            l = len(arr)
        tree = RBTree()
        for i in arr:
            tree.insert(i)
        lisb = []
        lt = 20
        for i in range(lt):
            ele = random.randint(0, s-1)
            ele = arr[ele]
            start = perf_counter_ns()
            val = tree.searchTree(tree.root, ele)
            end = perf_counter_ns()
            lisb.append(end-start)
            print(val)
            print("The time taken to search using rbt search is : {}".format(end-start))
        print("Average run time for size {} is : {}".format(s, sum(lisb)/lt))
        print("-------------------------------------------------------------------------------------")
        rbt.append(sum(lisb)/lt)

```
